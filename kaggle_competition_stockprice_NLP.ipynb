```python
from datetime import datetime

data.Date = data.Date.apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
```

```python
data.sort_values(by = 'Date', ascending = True, inplace = True)
```

#### Cleaning the Data Set

```python
import re
from unidecode import unidecode

def preprocessing_text(paragraph):
    
    paragraph = str(paragraph)
    para_clean = unidecode(paragraph)
    lower_case = para_clean.lower()
    no_numbers = re.sub(r'\d', '', lower_case) #Remove Numbers out of the Text
    no_special_chars = re.sub(r'[^ a-zA-Z]', '', no_numbers) #Remove Special Characters
    
    return no_special_chars

def select_columns(data, remove_cols):
    
    list_columns = list(data.columns)
    list_columns = [column for column in list_columns if column not in remove_cols]

    for column in list_columns:

        data[column] = data[column].apply(preprocessing_text)

    return data


def concatenate_strings(row):
    text = ' '.join(row)
    return (text + ' ')

data = select_columns(data, ['Date','Label'])
```

```python
data['consolidated_texts'] = data.drop(['Date', 'Label'], axis =1).apply(concatenate_strings, axis =1)
```

#### Separate Train and Test Data

```python
train = data[data['Date'] <= '20150101']
test = data[data['Date']  > '20141231']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
countvector = CountVectorizer(ngram_range = (2,2))
traindataset = countvector.fit_transform(train.consolidated_texts)
```

```python
randomclassifier = RandomForestClassifier(n_estimators = 200, criterion = 'entropy')
randomclassifier.fit(traindataset, train['Label']) 
```

#### Classification Metrics

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
testdataset = countvector.transform(test.consolidated_texts)
```

```python
predictions = randomclassifier.predict(testdataset)
```

```python
print(classification_report(test['Label'], predictions))
```
